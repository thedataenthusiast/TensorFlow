```python
import numpy as np
import tensorflow as tf
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris_df = pd.read_csv('iris.csv')
```

```python
iris_df = iris_df.sample(frac=1.0).reset_index(drop=True)
iris_df.head(10)
```

```python
iris_features_df = iris_df.drop(['species'], axis=1)
iris_features_df.head()
```

```python
iris_labels = iris_df['species'].values
iris_features = iris_features_df.values.astype(np.float32)
```

```python
iris_features
```

```python
def input_fn(iris_features):
    return tf.train.limit_epochs(
    tf.convert_to_tensor(iris_features, dtype=tf.float32), num_epochs=2)
```

```python
from tensorflow.contrib.factorization import KMeansClustering
kmeans_estimator = KMeansClustering(num_clusters=3, use_mini_batch=False)
```

```python
kmeans_estimator.train(input_fn=lambda: input_fn(iris_features))
# 166.5 is the square distance of all the data points from the cluster center
```

```python
cluster_centers = kmeans_estimator.cluster_centers()
print(cluster_centers)
```

```python
test_x = iris_features[100:120]
test_y = iris_labels[100:120]
cluster_indices = list(kmeans_estimator.predict_cluster_index(input_fn=lambda:input_fn(test_x)))

for i, point in enumerate(test_x):
    
    cluster_index = cluster_indices[i]
    center = cluster_centers[cluster_index]

    print("Flower with features:", point, ' is in cluster', cluster_index, ' label is', test_y[i])
    
```
